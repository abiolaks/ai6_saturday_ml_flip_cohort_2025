## 3. Trend Analysis

convert the year_built and tx_year to datetime format and understand the relationship with the transaction price over the years

```python
df.dtypes
```

```python
# Convert 'year_built' and 'year built' columns to datetime (YYYY-01-01)
for col in ['year_built', 'year built']:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col].astype(str), format='%Y', errors='coerce')
```

**Trend Analysis of Property Year Features**
Now that `year_built` and `tx_year` are in datetime format, we can analyze trends in property construction and transaction activity over time. The following plots will help visualize how the number of properties built and transactions have changed by year.

```python

# --- Time Series Trends (if datetime and target exist) ---
def find_main_datetime_col(cols):
    priority = ['year_built', 'year built']
    lower_map = {c.lower(): c for c in cols}
    for key in priority:
        for lc, orig in lower_map.items():
            if key in lc:
                return orig
    return cols[0] if cols else None

datetime_cols=['year_built', 'year built']

main_dt = find_main_datetime_col(datetime_cols)
target_col = "tx_price"

if (main_dt is not None) and (target_col is not None) and (target_col in df.columns):
    tmp = df[[main_dt, target_col]].copy().dropna(subset=[main_dt])
    if not pd.api.types.is_datetime64_any_dtype(tmp[main_dt]):
        tmp[main_dt] = pd.to_datetime(tmp[main_dt], errors="coerce")
    tmp = tmp.dropna(subset=[main_dt])
    if not tmp.empty:
        tmp["year"] = tmp[main_dt].dt.to_period("M").dt.to_timestamp()
        yearly = tmp.groupby("year").agg(
            Volume=(target_col, "count"),
            MedianTarget=(target_col, "median")
        ).reset_index()
        display(yearly.head(36))

        plt.figure()
        plt.plot(yearly["year"], yearly["Volume"])
        plt.title("Yearly Transaction Volume")
        plt.xlabel("year")
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig('yearly_transaction_volume.png') 
        plt.show()

else:
    print("No suitable datetime/target combination found for time trends.")
```

**Key patterns**

Pre-1930: almost no observations  explains why the avg price by year built was very noisy for the oldest years.

1940s1970s: steady rise in counts as building activity ramps up; occasional spikes/dips from small samples.

Mid-1980searly 2000s: peak volume (4060+ per year)lots of homes from these vintages, so price estimates there are more reliable.

Sharp drop after ~20052015: likely a data cutoff / right-censoring (few very recent builds in the dataset) and/or a real post-2008 slowdown.

```python
# Line plot of average transaction price by year_built
if 'year_built' in df.columns and 'tx_price' in df.columns:
    # Ensure year_built is datetime
    if not pd.api.types.is_datetime64_any_dtype(df['year_built']):
        df['year_built'] = pd.to_datetime(df['year_built'], errors='coerce')
    tmp = df.dropna(subset=['year_built', 'tx_price']).copy()
    tmp['year'] = tmp['year_built'].dt.year
    avg_price_by_year = tmp.groupby('year')['tx_price'].mean()
    plt.figure(figsize=(12, 6))
    plt.plot(avg_price_by_year.index, avg_price_by_year.values, marker='o', color='teal')
    plt.title('Average Transaction Price by Year Built')
    plt.xlabel('Year Built')
    plt.ylabel('Average Transaction Price')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('avg_tx_price_by_year_built.png')  # Save the plot as a PNG file
    plt.show()
else:
    print('year_built or tx_price column not found.')
```

**Big takeaways**

Pre-1950s = very volatile. Large spikes (e.g., ~1890, ~1910, mid-1940s) and dips occur because there are likely very few sales per build-year. With tiny sample sizes, one luxury or distressed sale can swing the average a lot.

19551985 = stable plateau. Averages hover roughly in the $350k$450k band with modest wigglessuggesting more consistent stock (lots of mid-century homes).

19852010 = gradual rise. A slow upward trend toward $450k$500k.

A sharp spike near the end (~20152016). That ~$700k point is almost certainly an outlier year (few observations or a cluster of high-end sales).
