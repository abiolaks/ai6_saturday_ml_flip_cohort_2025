```python
# NumPy for numerical computing
import numpy as np

# Pandas for DataFrames
import pandas as pd
pd.set_option('display.max_columns', 100)

# Matplotlib for visualization
from matplotlib import pyplot as plt
# display plots in the notebook
%matplotlib inline 

# Seaborn for easier visualization
import seaborn as sns
sns.set_style('darkgrid')
```

```python
# Load real estate data from CSV
df = pd.read_csv('../data/real_estate_data.csv')
```

```python
# Display the first 5 rows
df.head()
```

```python
# Get a concise summary of the DataFrame
df.info()
```

```python
# Get basic statistics for numeric columns
df.describe()
```

```python
# checking for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# chekcing the uniques values in the columns with msissing values
missing_values = missing_values[missing_values > 0]
for col in missing_values.index:
    print(f"Column: {col}")
    print(df[col].unique())
    # print the number of unique values
    print(f"Number of unique values: {df[col].nunique()}")
    print("\n")
```

```python
# Plot histogram  grid for the numerical columns at glance
df.hist(figsize=(14,14), xrot=-45)

# Clear the text "residue"
plt.show()
```

```python
# List of numerical columns to visualize
numerical_cols = ['tx_price', 'sqft', 'lot_size', 'property_tax', 'insurance']

# Plot histograms for key numerical features
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Plot box plots to identify outliers
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Box plot of {col}')
    plt.xlabel(col)
    plt.show()
```

**From the Box Plot** It shows that the following features has outlier which much examine and deal with.
* insurance
* sqft
* property_tax
* lot_size

**Also from the Historgram(KDE) plot** shows that most of the numerical feature are skew to the right hence the need for some transformation to address this issue.

```python

# List of categorical columns
categorical_cols = ['property_type', 'exterior_walls', 'roof']

# Plot bar charts for key categorical features
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
# Plotting the Pair Plot for key numerical features
sns.set(style="whitegrid")
numerical_features = ['tx_price', 'sqft', 'lot_size', 'property_tax']
pair_plot = sns.pairplot(df[numerical_features])
pair_plot.fig.suptitle('Pair Plot of Key Numerical Features', y=1.02, fontsize=16)
pair_plot.savefig('pair_plot.png')
plt.show()
```

```python
# Recreate the bubble chart with 'property_type' as hue and increased bubble size
plt.figure(figsize=(14, 10))
sns.scatterplot(
    x='sqft',
    y='tx_price',
    hue='property_type',
    size='lot_size',
    data=df,
    sizes=(50, 500),  # Increased bubble size for better definition
    alpha=0.6,
    palette='viridis'
)

# Set the plot title and labels
plt.title('Transaction Price vs. Square Footage, Sized by Lot Size and Colored by Property Type', fontsize=16)
plt.xlabel('Square Footage (sqft)', fontsize=12)
plt.ylabel('Transaction Price (tx_price)', fontsize=12)
plt.legend(title='Property Type')
plt.grid(True)
plt.tight_layout()

# Save the plot to a file
plt.savefig('larger_bubble_chart.png')
plt.show()

print("The new bubble chart with larger bubbles has been generated and saved as larger_bubble_chart.png.")
```

# PLot Explanation and Overview
## Insights into the real estate dataset.

### 1. Violin Plots
A violin plot is a powerful tool for exploring the distribution of a numerical variable across different categories. Unlike a standard box plot, it also displays the density of the data at different values, giving you a richer view of the data's shape.

**Key Insights**
Price Distribution by Property Type: The violin plot for tx_price and property_type (shown above) clearly illustrates the price distribution for apartments/condos, single-family homes, and townhouses. You can see the median price (the white dot), the interquartile range (the thick black bar), and the overall density of prices (the shape of the violin).
This plot shows that single-family homes tend to have a wider price range and a higher median price compared to apartments.

### 2. Pair Plots
A pair plot (or scatterplot matrix) is an excellent way to visualize the relationships between multiple numerical variables at once. It creates a grid of scatter plots for each pair of variables and histograms for each individual variable on the diagonal.

Key Insights
Correlation at a Glance: The pair plot of key numerical features (tx_price, sqft, lot_size, and property_tax) allows you to quickly identify potential correlations.

* **The plot of tx_price vs. sqft** shows a clear positive linear relationship, meaning larger properties tend to sell for higher prices.

* **The plot of tx_price vs. property_tax** also reveals a strong positive correlation, which makes sense as property taxes are often based on the assessed value.

**Distribution Shape**: The histograms on the diagonal provide a quick look at the distribution of each variable. For instance, you can see if a variable like sqft is normally distributed or skewed.

### 3. Bubble Charts
A bubble chart is a variation of a scatter plot where a third variable is represented by the size of the points (the "bubbles"). This is a great way to add another layer of information to a bivariate plot.

Key Insights
Triple-Variable Relationships: The bubble chart of tx_price versus sqft with bubble size determined by lot_size helps uncover a three-way relationship.

You can see the general trend that properties with larger square footage command higher prices.

By observing the bubble size, you can see if properties with a larger lot size (lot_size) also tend to have a higher price, even for a similar square footage. This helps to confirm that lot_size is an important factor in the transaction price.

Summary of Insights
Through these comprehensive visualizations and the initial EDA, we have uncovered several key insights about the real estate market:

Price is strongly driven by property size: There is a clear positive correlation between tx_price, sqft, and property_tax.

Single-family homes are the most valuable property type: The price distribution for single-family homes is higher and wider than for other property types.

Lot size also plays a significant role: Properties with larger lots tend to have higher prices, independent of their square footage.

Missing data patterns: Missing values in basement, lot_size.
